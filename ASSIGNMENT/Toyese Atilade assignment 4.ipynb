```python
import pandas as pd
import numpy as np

# Read the CSV file into a Pandas DataFrame
movies_data = pd.read_csv('movies.csv')

# Display the DataFrame
print(movies_data)
```

```python
# inspect the last rows

# Get the last 7 rows of the DataFrame
last_7_rows = movies_data.tail(7)

# Print the last 7 rows
print(last_7_rows)
```

```python
# check shape of the dataset
# Get the shape of the DataFrame (rows, columns)
shape = movies_data.shape

# Print the shape
print(shape)
```

```python
# check some specific rows of data

# Select rows 5000 to 5099 (inclusive)
selected_rows = movies_data.iloc[5000:5100]

# Print the selected rows
print(selected_rows)
```

```python
# inspect data types

# Get information about the DataFrame
movies_data.info()
```

```python
# check missing values

# Check for missing values in each cell
missing_values = movies_data.isnull()

# Print the DataFrame with True/False indicating missing values
print(missing_values)
```

```python
# check missing values

# Calculate the number of missing values in each column
missing_values = movies_data.isnull().sum()

# Print the missing values summary
print(missing_values)
```

```python
# Calculate descriptive statistics for numerical columns
numerical_stats = movies_data.describe()

# Print the descriptive statistics
print(numerical_stats)
```

```python
# Calculate the percentage of gross revenue to budget
movies_data['budget_to_gross_percentage'] = (movies_data['gross'] / movies_data['budget']) * 100

# Format the percentage values as strings with 2 decimal places and '%'
movies_data['budget_to_gross_percentage'] = movies_data['budget_to_gross_percentage'].apply(lambda x: f'{x:.2f}%')

# Display the DataFrame
print(movies_data)
```

```python
# Create a new column 'good_movie' based on the 'score' column
movies_data['good_movie'] = ['Yes' if x > 8.0 else 'No' for x in movies_data['score']]

# Display the first 5 rows of the modified DataFrame
print(movies_data.head())
```

```python
# Filter movies with 'good_movie' == 'yes'
good_movies = movies_data[movies_data['good_movie'] == 'yes']

# Display the first 3 rows of the filtered DataFrame
print(good_movies.head(3))
```

DATA VISUALIZATION

```python

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set_style('whitegrid')
sns.set_palette('dark')
```

```python
# Sample columns for each type of plot
# Define numerical and categorical columns
num_cols = ['score', 'votes', 'budget', 'gross', 'runtime']
cat_cols = ['genre', 'rating', 'country']
```

BAR CHART

```python
# Generate 3 Bar Charts
# Select the first 3 categorical columns
cat_cols_to_plot = cat_cols[:3]

# Create a figure with subplots (3 rows, 1 column)
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

# Create bar plots for each categorical column
for i, col in enumerate(cat_cols_to_plot):
    sns.countplot(data=movies_data, x=col, order=movies_data[col].value_counts().index, ax=axes[i])
    axes[i].set_title(f"Bar Chart of {col.capitalize()}")
    plt.xticks(rotation=45)

# Adjust layout to prevent overlapping elements
plt.tight_layout()

plt.show()
```

HISTOGRAM

```python
# Generate 3 Histograms
# Select the first 3 numerical columns
num_cols_to_plot = num_cols[:3]

# Create a figure with subplots (3 rows, 1 column)
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

# Create histograms for each column
for i, col in enumerate(num_cols_to_plot):
    sns.histplot(data=movies_data, x=col, kde=True, bins=30, ax=axes[i])
    axes[i].set_title(f"Histogram of {col.capitalize()}")

# Adjust layout to prevent overlapping elements
plt.tight_layout()

plt.show()
```

SCATTER PLOT

```python
# Generate 3 Scatter Plots
# Create the scatter plot with color-coded points
sns.scatterplot(data=movies_data, x='votes', y='score', hue='rating')

plt.title("Movie Score vs. Votes by Rating")
plt.xlabel("Votes")
plt.ylabel("Score")

plt.show()
```

BOXPLOT

```python
# Assuming 'num_cols' is a list containing numerical columns

# Select the first 3 numerical columns
num_cols_to_plot = num_cols[:3]

# Create a figure with subplots (3 rows, 1 column)
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

# Create box plots for each column
for i, col in enumerate(num_cols_to_plot):
    sns.boxplot(ax=axes[i], data=movies_data, x=col)
    axes[i].set_title(f"Box Plot of {col.capitalize()}")

# Adjust layout to prevent overlapping elements
plt.tight_layout()

plt.show()
```

LINEPLOT

```python
# Create the line plot
sns.lineplot(data=movies_data, x='score', y='votes')

# Customize the plot
plt.title("Votes vs. Score")
plt.xlabel("Score")
plt.ylabel("Votes")

plt.show()
```

VIOLIN PLOT

```python
# Create the violin plot
sns.violinplot(data=movies_data, x='score', inner="quartile")

# Customize the plot
plt.title("Distribution of Movie Scores")
plt.xlabel("Score")
plt.ylabel("Density")

plt.show()
```

HEATMAP

```python
subset_data = movies_data[['score', 'budget']]

# Calculate correlation matrix
correlation_matrix = subset_data.corr()

# Create a heatmap with customizations
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Score and Budget")
plt.show()
```
