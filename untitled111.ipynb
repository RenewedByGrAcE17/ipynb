```python
# create categorical and continuous column
movies_df.columns
cat_cols = ['name', 'rating', 'genre','released',
       'director', 'writer', 'star', 'country', 'company']
cont_cols = [x for x in list(movies_df.columns) if x not in cat_cols]

for col in cat_cols:
  movies_df[col] = movies_df[col].fillna('unknown')

for col in cont_cols:
  movies_df[col] = movies_df[col].fillna(movies_df[col].median())
```

### INFERENTIAL STATISTICS

```python
# steps in hypothesis testing

# set up the hypothesis
ho = 'There is no significant difference between gross across ratings'
h1 = 'There is a significant difference between gross across ratings'

# choose level of significance
alpha = 0.05

# setup the decision rule
def decision(p_value:float):

  if p_value < alpha:
    result = h1
  else:
    result = ho
  return f'{result} @ pvalue: {round(p_value, 4)}'

# compute the test
model = ols(formula = 'gross ~ rating', data=movies_df).fit()
anova_table = anova_lm(model)
rating_pvalue = anova_table['PR(>F)'].loc['rating']

# get conclusion
decision(p_value=rating_pvalue)
```
